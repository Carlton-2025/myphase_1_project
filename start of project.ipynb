```python
aviation_df.drop_duplicates(inplace=True)
aviation_df
```

```python
aviation_df.drop(columns=['Latitude', 'Longitude','Airport.Code','Amateur.Built','FAR.Description','Airport.Name','Engine.Type','Number.of.Engines','Aircraft.Category','Air.carrier','Broad.phase.of.flight','Weather.Condition','Report.Status','Schedule'], inplace=True)
aviation_df
```

```python
missing_values_injury_severity= aviation_df['Injury.Severity'].isnull().sum()
missing_values_aircraft_damage= aviation_df['Aircraft.damage'].isnull().sum()
missing_values_purpose_of_flight= aviation_df['Purpose.of.flight'].isnull().sum()
missing_values_fatal_injury= aviation_df['Total.Fatal.Injuries'].isnull().sum()
missing_values_serious_injury= aviation_df['Total.Serious.Injuries'].isnull().sum()
missing_values_minor_injury= aviation_df['Total.Minor.Injuries'].isnull().sum()
print(missing_values_injury_severity)
print(missing_values_aircraft_damage)
print(missing_values_purpose_of_flight)
print(missing_values_fatal_injury)
print(missing_values_serious_injury)
print(missing_values_minor_injury)
```

```python
aviation_df = aviation_df.dropna(subset=['Injury.Severity'])
aviation_df
```

```python
aviation_df = aviation_df.dropna(subset=['Aircraft.damage'])
aviation_df 
```

```python
aviation_df = aviation_df.dropna(subset=['Purpose.of.flight'])
aviation_df 
```

```python

mode_value_fatal_injuries = aviation_df['Total.Fatal.Injuries'].mode()[0]
aviation_df.loc[:,'Total.Fatal.Injuries']=aviation_df['Total.Fatal.Injuries'].fillna(mode_value_fatal_injuries)
aviation_df 
```

```python
 
mode_value_serious_injuries = aviation_df['Total.Serious.Injuries'].mode()[0]
aviation_df.loc[:,'Total.Serious.Injuries'] =aviation_df['Total.Serious.Injuries'].fillna(mode_value_serious_injuries)
aviation_df 
```

```python
mode_value_minor_injuries = aviation_df['Total.Minor.Injuries'].mode()[0]
aviation_df.loc[:,'Total.Minor.Injuries']=aviation_df['Total.Minor.Injuries'] .fillna(mode_value_minor_injuries )
aviation_df 
```

```python
mean_value_uninjured = aviation_df['Total.Uninjured'].mean()
aviation_df.loc[:,'Total.Uninjured']=aviation_df['Total.Uninjured'] .fillna(mean_value_uninjured )
aviation_df 
```

```python

aviation_df.loc[:, 'Publication.Date'] = aviation_df['Publication.Date'].ffill()
aviation_df
```
